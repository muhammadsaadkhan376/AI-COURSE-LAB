```python
# Task 2: Dictionary Operations

# Step 1: Create a dictionary
student = {
    "name": "saad",
    "age": 20,
    "grade": "A",
    "email": "munazza@example.com"
}

# Step 2: Access and print values
print("Student Name:", student["name"])
print("Student Age:", student["age"])

# Step 3: Modify a value
student["grade"] = "A+"
print("Updated Grade:", student["grade"])

# Step 4: Add a new key-value pair
student["major"] = "Computer Science"
print("Student Major:", student["major"])

# Step 5: Remove a key-value pair
del student["email"]
print("Student dictionary after removing email:", student)

# Step 6: Iterate through keys and values
print("All Keys:")
for key in student.keys():
    print(key)

print("All Values:")
for value in student.values():
    print(value)

print("All Items:")
for key, value in student.items():
    print(f"{key}: {value}")
```
